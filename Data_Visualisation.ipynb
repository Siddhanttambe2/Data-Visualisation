```python
#Seaborn Assignment
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
#Q1) 1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.
```

```python
np.random.seed(42)
n = 100
x = np.random.normal(loc=0, scale=1, size=n)
y = 2 * x + np.random.normal(loc=0, scale=1, size=n)
df = pd.DataFrame({'X': x, 'Y': y})
sns.scatterplot(data=df, x='X', y='Y')
plt.title('Scatter Plot of Synthetic Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#Q2) Generate a dataset of random numbers. Visualize the distribution of a numerical variable
```

```python
np.random.seed(42)
n = 1000
data = np.random.normal(loc=0, scale=1, size=n)
df = pd.DataFrame({'Value': data})
sns.histplot(df['Value'], kde=True, bins=30)
plt.title('Distribution of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#Q3) Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.
```

```python
np.random.seed(42)
categories = ['A', 'B', 'C', 'D']
values = np.random.normal(loc=0, scale=1, size=(100, len(categories)))
df = pd.DataFrame(values, columns=categories)
df_melted = df.melt(var_name='Category', value_name='Value')
plt.figure(figsize=(10, 6))
sns.barplot(data=df_melted, x='Category', y='Value', estimator='mean', ci=None)
plt.title('Comparison of Categories Based on Values')
plt.xlabel('Category')
plt.ylabel('Mean Value')
plt.show()
```

```python
#Q4) Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories
```

```python
np.random.seed(42)
categories = ['A', 'B', 'C', 'D']
values = np.random.normal(loc=0, scale=1, size=(100, len(categories)))
df = pd.DataFrame(values, columns=categories)
df_melted = df.melt(var_name='Category', value_name='Value')

sns.boxplot(x='Category', y='Value', data=df_melted)
plt.title('Distribution of Numerical Variable Across Categories')
plt.show()
```

```python
#Q5)Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
```

```python
np.random.seed(42)
n = 100
x1 = np.random.normal(loc=0, scale=1, size=n)
x2 = 0.8 * x1 + np.random.normal(loc=0, scale=0.5, size=n)
x3 = 0.3 * x1 + 0.5 * x2 + np.random.normal(loc=0, scale=0.5, size=n)
df = pd.DataFrame({'Feature1': x1, 'Feature2': x2, 'Feature3': x3})
corr_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Features')
plt.show()
```

```python
#Q3) Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
```

```python
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
pivot_table = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')
fig = px.imshow(pivot_table,
                labels={'x': 'Month', 'y': 'Day', 'color': 'Sales'},
                color_continuous_scale='YlGnBu',
                title='Variation in Sales Across Different Months and Days')
fig.show()
```

```python
#Q4) Using the given x and y data, generate a 3D surface plot to visualize the function
```

```python

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(
    title='3D Surface Plot of z = sin(x² + y²)',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)
fig.show()
```

```python
#BOKEH ASSIGNMENT
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral11
from bokeh.palettes import Spectral4
import numpy as np
import pandas as pd
```

```python
#Q1) Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

output_notebook()
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)
p.line(x, y, line_width=2, color='blue', legend_label='sin(x)')
show(p)
```

```python
#Q2) Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
```

```python

np.random.seed(42)
n = 100
x = np.random.uniform(0, 10, n)
y = np.random.uniform(0, 10, n)
sizes = np.random.uniform(10, 50, n)
colors = np.random.choice(Spectral11, n)
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))
output_notebook()
p = figure(title="Scatter Plot with Varying Sizes and Colors", x_axis_label='X', y_axis_label='Y', width=800, height=400)
p.scatter('x', 'y', size='sizes', color='colors', alpha=0.6, legend_field='colors', source=source)
p.legend.title = 'Colors'
show(p)
```

```python
 #Q3) Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
```

```python
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
output_notebook()
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count',
           width=800, height=400, toolbar_location=None)

p.vbar(x=fruits, top=counts, width=0.5, color=Spectral4[0], legend_label="Count")
p.xaxis.major_label_orientation = "vertical"
p.legend.orientation = "horizontal"
p.legend.location = "top_left"
show(p)
```

```python
#Q4) Create a Bokeh histogram to visualize the distribution of the given data.
```

```python
np.random.seed(42)
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
output_notebook()
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency',
           width=800, height=400)
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='skyblue', line_color='black')
show(p)
```

```python
#Q5)  Create a Bokeh heatmap using the provided dataset.
```

```python
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
from bokeh.models import ColorBar, ColumnDataSource # Import ColumnDataSource
from bokeh.layouts import column


data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

output_notebook()

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           width=600, height=600, toolbar_location=None,
           tools="hover", tooltips=[("Value", "@z")])

mapper = linear_cmap(field_name='z', palette=Viridis256, low=min(z_flat), high=max(z_flat))

source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, z=z_flat))

p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
       fill_color=mapper, line_color=None)

color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')
show(p)
```
